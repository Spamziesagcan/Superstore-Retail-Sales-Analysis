REMOVING OUTLIERS

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

DATA VISUALIZATION:

1. HISTOGRAM

```python
sns.histplot(df['Category'],kde=False,bins=50)
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

2. PIE CHART

```python
sales_by_category = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(30,12))
plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Category')
plt.show()
```

3. SCATTER PLOT

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["Quantity"], df["Profit"], alpha=0.5, color="cyan")
plt.xlabel("Quantity Sold")
plt.ylabel("Profit ($)")
plt.title("Profit vs Quantity")
plt.grid(True)
plt.show()
```

4. BAR CHART

```python
sales_by_ship_mode = df.groupby("Ship Mode")["Sales"].sum()
plt.figure(figsize=(8, 6))
plt.bar(sales_by_ship_mode.index, sales_by_ship_mode, color=["blue", "green", "red", "purple"])
plt.xlabel("Ship Mode")
plt.ylabel("Total Sales ($)")
plt.title("Sales by Ship Mode")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

5. BOX PLOT

```python
plt.figure(figsize=(20, 100))
sns.boxplot(x="Region", y="Sales", data=df, palette="pastel")
plt.xlabel("Region")
plt.ylabel("Sales ($)")
plt.title("Sales Distribution by Region")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

6. HEATMAP

```python
sales_by_city_category = df.pivot_table(values="Sales", index="Region", columns="Category", aggfunc="sum")
plt.figure(figsize=(10, 8))
sns.heatmap(sales_by_city_category, annot=True, cmap="Blues", linewidths=0.5, fmt=".0f")
plt.title("Sales by Region and Category")
plt.xlabel("Category")
plt.ylabel("Region")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

7. 3D SCATTER PLOT

```python
fig = plt.figure(figsize=(15, 9))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["Sales"], df["Profit"], df["Quantity"], c=df["Profit"], cmap="Blues", alpha=0.7)
ax.set_xlabel("Sales ($)")
ax.set_ylabel("Profit ($)")
ax.set_zlabel("Quantity Sold")
ax.set_title("3D Scatter Plot: Sales vs. Profit vs. Quantity")
plt.show()
```

8. VIOLIN PLOT

```python
plt.figure(figsize=(15, 9))
sns.violinplot(x="Profit", y="Region", data=df, palette="pastel", inner="box")
plt.xlabel("Profit ($)")
plt.ylabel("Region")
plt.title("Profit Distribution by Region")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

9. PAIR PLOT

```python
num_cols = ["Sales", "Profit", "Quantity"]
sns.pairplot(df, vars=num_cols, hue="Category", diag_kind="kde", palette="bright", plot_kws={"alpha": 0.7})
plt.show()
```

10. KDE PLOT

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x="Sales", hue="Region", fill=True, common_norm=False, palette="Set2", linewidth=2)
plt.xlabel("Sales ($)")
plt.ylabel("Density")
plt.title("Sales Distribution by Region")
plt.xlim(0, df["Sales"].quantile(0.95))
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

11. CATPLOT

```python
sns.catplot(x="Category", y="Sales", col="Region", data=df, kind="bar", palette="pastel", height=4, aspect=1)
plt.subplots_adjust(top=0.85)
plt.suptitle("Sales by Category Across Regions")
plt.show()
```
